# UFUNCS

**UNIVERSAL FUNCTIONS ARE NUMPY FUNCTION THAT ARE USED TO OPERATE *ndarray* OBJECTS**  
**UFUNCS ARE USED TO IMPLEMENT VECTORIZATION (A FASTER WAY TO ITERATE OVER ELEMENTS)**  
**CONVERTING ITERATIVE STATEMENTS INTO VECTOR BASED OPERATION IS CALLED VECTORIZATION**

```python
x = [1,2,3,4]
y = [5,6,7,8]
z = []
for i,j in zip(x,y):
    z.append(i+j)
z
```

```python
np.add(x,y)
```

## CREATING OWN NUMPY UFUNC

```python
def mymul(a,b):
    return a*b
```

```python
def square(z):
    return z*z
```

```python
vfunc = np.vectorize(square)
```

```python
vfunc(2)
```

```python
m = np.frompyfunc(mymul,2,1)
m([1,2,3],[4,5,6])
```

```python
type(m)
```

```python
type(np.add)
```
