```python
def Check_Prime(n):
    i = 2
    flag = 0
    while i<n:
        if n%i == 0:
            flag = 1
        i+=1
    if flag == 1:
        return 'NOT PRIME'
    else:
        return 'PRIME'
```

```python
Check_Prime(7)
```

```python
df['Prime/Not'] = df.Ratings.apply(Check_Prime)
```
