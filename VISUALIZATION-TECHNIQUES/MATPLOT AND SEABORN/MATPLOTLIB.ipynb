# MATPLOTLIB PLOTS

```python
import matplotlib
matplotlib.__version__
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd
```

```python
iris = load_iris()
```

```python
df = pd.DataFrame(iris.data,columns = iris.feature_names)
df['target'] = iris.target
```

```python
df
```

```python
df.head()
```

```python
x = df.iloc[:5,0]#sepal_length
y = df.iloc[:5,1]#sepal_width
```

```python
x
```

```python
y #(2,3)
```

## PLOTS

```python
plt.plot(x,y)
```

**PLOT THE X-AXIS WITH CORRESPONDING Y-AXIS**

```python
plt.plot(x,y,'4') # use : x, +,^,*,s,P,H,h,_,|,1:4
plt.show()
```

**USE DIFFERENT STYLE OF POINT MARKERS**

```python
plt.plot(x)
```

**IF ONE OF THE AXIS IS NOT SPECIFIED IT WILL ASSUME THE CO-ORDINATE STARTS FROM 0,1,2....**

```python
plt.plot(x,y,marker = 's')
plt.show()
```

```python
plt.plot(x,y,'*-.k') # marker|line|color
plt.show()
```

```python
plt.plot(x,y,'o--y',ms = 20)#marker_size
```

```python
plt.plot(x,y,'o--y',ms = 20,mec = 'k') #marker_edge_color
```

```python
plt.plot(x,y,'o--y',ms = 20,mec = 'g',mfc = 'b') #marker_face_color
```

```python
plt.plot(x,y,ls = ':') #linestyle
```

```python
plt.plot(x,y,ls = ':',c = 'hotpink')
```

```python
plt.plot(x,y,marker = 'o',ls = '-.',c = 'k',linewidth = 5,ms = 20)
```

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])
```

```python
plt.plot(x1,y1,label = 'plot1')
plt.plot(x2,y2,label = 'plot2')
plt.legend()
```

```python
plt.plot(x1,y1,x2,y2)
```

```python
plt.plot(x,y)
font1 = {'family':'serif','color':'blue','size':20}
plt.title('sepal length vs width',loc = 'right',fontdict = font1)
plt.xlabel('Sepal length')
plt.ylabel('sepal width')
```

## GRID

```python
plt.plot(x,y)
font1 = {'family':'serif','color':'blue','size':20}
plt.title('sepal length vs width',loc = 'right',fontdict = font1)
plt.xlabel('Sepal length')
plt.ylabel('sepal width')
plt.grid()
```

```python
plt.plot(x,y)
font1 = {'family':'serif','color':'blue','size':20}
plt.title('sepal length vs width',loc = 'right',fontdict = font1)
plt.xlabel('Sepal length')
plt.ylabel('sepal width')
plt.grid(axis = 'x')# try axis = y
```

```python
plt.plot(x,y)
font1 = {'family':'serif','color':'blue','size':20}
plt.title('sepal length vs width',loc = 'right',fontdict = font1)
plt.xlabel('Sepal length')
plt.ylabel('sepal width')
plt.grid(ls = '-.',color = 'k',linewidth = 0.9)
```

## SUBPLOTS

```python
plt.subplot(1,3,1)# rows|column|position
plt.plot(x,y)

plt.subplot(1,3,2)
plt.plot(x1,y1)

plt.subplot(1,3,3)
plt.plot(x2,y2)
```

```python
plt.subplot(3,1,1)
plt.plot(x,y)
plt.title('subplot 1')

plt.subplot(3,1,2)
plt.plot(x1,y1)
plt.title('subplot 2')

plt.subplot(3,1,3)
plt.plot(x2,y2)
plt.title('subplot 3')

plt.suptitle('SUBPLOTS')
plt.show()
```

## SCATTER

## BAR 

```python
names = np.array(['a','b','c','d','e'])
plt.bar(names,x,color = 'yellow',width = 0.7)
plt.show()
```

```python
names = np.array(['a','b','c','d','e'])
plt.barh(names,x,height = 0.1)
plt.show()
```
