```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
```

```python
X, y = make_moons(n_samples=300, noise=0.3, random_state=42)
```

```python
X
```

```python
set(y)
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.title("Original Data")
plt.show()
```

```python
dbscan = DBSCAN(eps=0.2, min_samples=5)
dbscan.fit(X)
# df[['age']]
```

```python
labels = dbscan.labels_
```

```python
labels
```

```python
set(labels)
```

```python
core_samples = dbscan.core_sample_indices_

print("Core Points Count:", len(core_samples))
print("Unique Cluster Labels:", set(labels))
```

```python
core_samples
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow')

plt.title("DBSCAN Clustering Result")
plt.show()
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow', alpha=0.5)
plt.scatter(X[core_samples, 0], 
            X[core_samples, 1], 
            c='black', 
            s=50, 
            label='Core Points')

plt.legend()
plt.title("Core Points Highlighted")
plt.show()
```

```python
dbscan = DBSCAN(eps=0.1, min_samples=5)
labels = dbscan.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow')
plt.title("DBSCAN with smaller eps")
plt.show()
```

```python
from sklearn.neighbors import NearestNeighbors
neighbors = NearestNeighbors(n_neighbors=5)
neighbors_fit = neighbors.fit(X)
distances, indices = neighbors_fit.kneighbors(X)
distances = np.sort(distances[:, 4])
plt.plot(distances)
plt.title("K-distance Graph")
plt.grid()
plt.show()
```
