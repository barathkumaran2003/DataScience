```python
import pandas as aju
from sklearn.datasets import load_iris,load_digits
from sklearn.model_selection import train_test_split,cross_val_score,GridSearchCV,RandomizedSearchCV
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression,LinearRegression
from sklearn.naive_bayes import GaussianNB,MultinomialNB
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
import numpy as np
```

```python
iris = load_iris()
#iris.data
```

```python
df = aju.DataFrame(iris.data,columns = iris.feature_names)
df
```

```python
df['target'] = iris.target
df.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.drop('target',axis= 'columns'),df.target,test_size = 0.33)
```

```python
model = SVC(C=40,kernel='rbf',gamma = 'auto')
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
kval = ['linear', 'rbf']
cval = [1,10,20]
avg_score = {}
for k in kval:
    for c in cval:
        crss = cross_val_score(SVC(kernel = k,C = c,gamma = 'auto'),iris.data,iris.target,cv = 5)
        avg_score[k+'_'+str(c)] = np.average(crss)   
```

```python
avg_score
```

```python
clf = GridSearchCV(SVC(gamma = 'auto'),{
    'C' : [1,10,20],
    'kernel' : ['rbf','linear']
}, return_train_score=False,cv = 5)
```

```python
clf.fit(iris.data,iris.target)
clf.cv_results_
```

```python
dt = aju.DataFrame(clf.cv_results_)
dt
```

```python
dt[['param_C','param_kernel','mean_test_score']]
```

```python
clf.best_params_
```

```python
clf.best_score_
```

```python
rs = RandomizedSearchCV(SVC(gamma='auto'),{
    'C':[1,10,20],
    'kernel':['rbf','linear']
},cv = 5,return_train_score = False,
                       n_iter = 3)
rs.fit(iris.data,iris.target)
rs_df = aju.DataFrame(rs.cv_results_)
rs_df[['param_C','param_kernel','mean_test_score']]
```

```python
model_params = {
    'svm':{
        'model': SVC(gamma = 'auto'),
        'params':{
            'C':[1,10,20],
            'kernel':['rbf','linear']
        }
    },
    'rfc':{
        'model': RandomForestClassifier(),
        'params':{
            'n_estimators':[1,5,10]
        }
    },
    'log':{
        'model':LogisticRegression(solver='liblinear',multi_class='auto'),
        'params':{
            'C':[1,5,10]
        }
    }
}
```

```python
for model_names,mp in model_params.items():
    #print(model_names)
    print(mp)
```

```python
score = []
for model_names,mp in model_params.items():
    gsc = GridSearchCV(mp['model'],mp['params'],return_train_score = False)
    gsc.fit(iris.data,iris.target)
    score.append({
        'model':model_names,'best_score':gsc.best_score_,'best_param':gsc.best_params_
    })
```

```python
score
```

```python
dfg = aju.DataFrame(score,columns = ['model','best_score','best_param'])
dfg
```

```python
digit = load_digits()
```

```python
model_params = {
    'svc':{
        'model':SVC(gamma = 'auto'),
        'params':{
            'C':[1,10,20],
            'kernel':['rbf','linear']
        }
    },
    'rfc':{
        'model':RandomForestClassifier(),
        'params':{
            'n_estimators':[1,5,10]
        }
    },
    'log':{
        'model':LogisticRegression(solver = 'liblinear',multi_class = 'auto'),
        'params':{
            'C':[1,5,10]
        }
    },
    'dcr':{
        'model':DecisionTreeRegressor(),
        'params':{
            'criterion':["squared_error", "friedman_mse", "absolute_error"],
            'max_depth':[None,5,3,7]
        }
    },
    'gnb':{
        'model':GaussianNB(),
        'params':{}
    },
    'mnb':{
        'model':MultinomialNB(),
        'params':{}
    },
    'dtc':{
        'model':DecisionTreeClassifier(),
        'params':{
            'criterion':['gini','entropy']
        }
    }
}
```

```python
scores = []
for model_name,mp in model_params.items():
    gsc = GridSearchCV(mp['model'],mp['params'],cv = 5,return_train_score = False)
    gsc.fit(digit.data,digit.target)
    scores.append({'model':model_name,'best_score':gsc.best_score_,'best_param':gsc.best_params_})
df = aju.DataFrame(scores,columns = ['model','best_score','best_param'])
```
